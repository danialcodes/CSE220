```python
#Task-1
def SSRecursive(array,i,minimum):
    if i==len(array):
        return array
    minimum=i        
    minimum=findmaximum(array,i+1,minimum)        
    if minimum!=i:
        temp = array[minimum]
        array[minimum] = array[i]
        array[i] = temp
    return SSRecursive(array,i+1,minimum)
def findmaximum(array,i,minimum):
    if i==len(array):
        return minimum
    if array[i]<array[minimum]:
        minimum=i
    return findmaximum(array,i+1,minimum)  

print('\nTask-01:')
arr1=[6,2,1,5,9,4,6,10,5,3,4] 
print(SSRecursive(arr1,0,0))


#Task-2
def insert(array,i,temp):
    if i<0:
        return i,array
    if temp>array[i]:
        return i,array
    array[i+1]=array[i]
    return insert(array,i-1,temp)                
def ISRecursive(array,i,j,temp):
    if i==len(array):
        return array
    temp=array[i]        
    j,array=insert(array,i-1,temp)
    array[j+1]=temp
    return ISRecursive(array,i+1,j,temp)


print('\nTask-02:')
arr2=[1,5,6,8,1,2,3,8,5,8,4,3,2,7]
print(ISRecursive(arr2,i=1,j=0,temp=0))


class sNode:
    def __init__(self,data,next):
        self.data=data
        self.next=next
class Llist:
    def __init__(self,arr):
        self.head=None
        tail=None
        for i in range(len(arr)):
            node=sNode(arr[i],None)
            if self.head is None:
                self.head=node
                tail=node
            else:
                tail.next=node
                tail=node
    def showlist(self):
        itr=self.head
        while itr is not None:
            if itr.next is None:
                print(str(itr.data))
            else:                
                print(str(itr.data)+'->',end='')
            itr=itr.next                           

    #Task-03
    def bubbleSort(self):
        itr=self.head
        while itr is not None:
            itr2=self.head
            while itr2.next:
                if itr2.data>itr2.next.data:
                    temp=itr2.data
                    itr2.data=itr2.next.data
                    itr2.next.data=temp
                itr2=itr2.next
            itr=itr.next
    #Task-04 
    def selectionSort(self):
        n = self.head
        while n is not None:
            minimum=n
            n2=n.next
            while n2:
                if n2.data<minimum.data:
                    minimum=n2
                n2=n2.next

            temp=minimum.data
            minimum.data=n.data
            n.data=temp     
            n=n.next

print('\nTask-03:')
li1=Llist([50,20,60,90,80,55,40,10,30])
li1.bubbleSort()
li1.showlist()
print('\nTask-04:')
li2=Llist([2,8,7,10,25,66,42,33,12,6,5,1])
li2.selectionSort()
li2.showlist()



#Task-05
class DNode:
    def __init__(self,data,next,prev):
        self.data=data
        self.next=next
        self.prev=prev
class DLList:
    def __init__(self,arr):
        self.head=None
        tail=None
        node=DNode(None,None,None)
        self.head=node
        tail=node
        for i in range(len(arr)):
            node=DNode(arr[i],None,None)
            tail.next=node
            node.prev=tail
            tail=node
    def showlist(self):
        itr=self.head.next
        while itr:
            if itr.next==None:
                print(str(itr.data))
            else:                
                print(str(itr.data)+'->',end='')
            itr=itr.next            

    def insertionSort(self):
        n=self.head.next.next
        while n:
            x=n.data
            n2=n.prev
            while n2!=self.head and x<n2.data: 
                n2.next.data=n2.data
                n2=n2.prev
            n2.next.data=x
            n=n.next

print('\nTask-05:')
li3=DLList([15,29,100,7,35,77,6,2])
li3.insertionSort()
li3.showlist()



#Task-06
def binarySearch(arr,find,j,i=0):
    mid=(i+j)//2
    if(i>j):
        return -1
    if arr[mid]==find:
        return mid
    if arr[mid]<find:
        i=mid
    if arr[mid]>find:
        j=mid
    return binarySearch(arr,find,j,i)


print('\nTask-06:')
arr3=[13,16,30,46,52,99,200]
print(binarySearch(arr3,16,len(arr3)))

#Task-07
m = [0]*100000
def fib(n):
    if(n<0):
        raise IndexError()
    if n==1 or n==2:
        m[n] = 1
    if m[n]>0:
        return m[n]
    else:
        m[n]=fib(n-1)+fib(n-2)
    return m[n]

print('\nTask-07:')
print(fib(5))
```
